```python
import pandas as pd
df = pd.read_csv('JC-202210-citibike-tripdata.csv')
```

```python
# Convert 'started_at' and 'ended_at' to datetime if not already in that format
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Extract only the date part
df['started_at'] = df['started_at'].dt.date
df['ended_at'] = df['ended_at'].dt.date

# Display the updated dataframe
df.head()
```

```python
df.to_csv('_202210-citibike-tripdata_cleaned_data.csv', index=False)
```
